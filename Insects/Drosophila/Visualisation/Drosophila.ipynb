Visualize Drosophila neurons in 3D using Python with the navis and flybrains packages. This script loads an example neuron from the hemibrain dataset and displays it inside a standard brain mesh (FAFB14) with interactive Plotly rendering, allowing rotation, zoom, and pan directly in Google Colab. Colors and transparency can be customized, and multiple neurons can be plotted simultaneously. This provides a lightweight, reproducible way to explore neuron morphology in the context of the whole fly brain.

```python
import navis
import flybrains

# Example neuron
neuron = navis.example_neurons(1)

# Brain mesh (pick a template that exists, e.g., FAFB14)
brain = flybrains.FAFB14

# Plot
navis.plot3d(
    [brain, neuron],
    backend="plotly",
    color=["lightgrey", "red"],
    alpha=[1.0]  # only one neuron, alpha list = 1
)
```
